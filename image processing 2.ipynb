```python
path="image3.jpg"
image = cv2.imread(path)
```

```python
plt.imshow(image)
plt.show()
```

```python
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        for each in range(3):            
            if image[i,j][each]<=200:
                image[i,j][each]=0
            else:
                image[i,j][each]=255
```
