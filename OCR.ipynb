```python
import pytesseract

import matplotlib.pyplot as plt

import cv2
```

```python
im = cv2.imread("maxresdefault.JPG", cv2.IMREAD_COLOR)
```

```python

im = cv2.bitwise_not(im)
plt.imshow(im)
plt.show()
```

```python
config = ('-l eng --oem 1 --psm 3')
```

```python
text = pytesseract.image_to_string(im, config=config)
```

```python
print(text)
```
